# Format data from BDI-Sen

```python
df_train = pd.read_csv("./train-with-severities-and-multilabels.csv")
df_valid = pd.read_csv("./val-with-severities-and-multilabels.csv")
df_test = pd.read_csv("./test-with-severities-and-multilabels.csv")
```

```python
columns_mapping = {
    'Sadness': "1_annot", 
    'Pessimism': "2_annot", 
    'Sense_of_failure': "3_annot", 
    'Loss_of_Pleasure': "4_annot", 
    'Guilty_feelings': "5_annot", 
    'Sense_of_punishment': "6_annot", 
    'Self-dislike': "7_annot", 
    'Self-incrimination': "8_annot",
    'Suicidal_ideas': "9_annot", 
    'Crying': "10_annot", 
    'Agitation': "11_annot", 
    'Social_withdrawal': "12_annot", 
    'Indecision': "13_annot", 
    'Feelings_of_worthlessness': "14_annot", 
    'Loss_of_energy': "15_annot",
    'Change_of_sleep': "16_annot", 
    'Irritability': "17_annot", 
    'Changes_in_appetite': "18_annot", 
    'Concentration_difficulty': "19_annot",
    'Tiredness_or_fatigue': "20_annot", 
    'Loss_of_interest_in_sex': "21_annot"
}
```

```python
type_mapping = {
    "1_annot": "int32", 
    "2_annot": "int32", 
    "3_annot": "int32", 
    "4_annot": "int32", 
    "5_annot": "int32", 
    "6_annot": "int32", 
    "7_annot": "int32", 
    "8_annot": "int32",
    "9_annot": "int32", 
    "10_annot": "int32", 
    "11_annot": "int32", 
    "12_annot": "int32", 
    "13_annot": "int32", 
    "14_annot": "int32", 
    "15_annot": "int32",
    "16_annot": "int32", 
    "17_annot": "int32", 
    "18_annot": "int32", 
    "19_annot": "int32",
    "20_annot": "int32", 
    "21_annot": "int32"
}
```

```python
df_train
```

```python
df_train = df_train.rename(columns=columns_mapping)
df_valid = df_valid.rename(columns=columns_mapping)
df_test = df_test.rename(columns=columns_mapping)
```

```python
df_train = df_train.astype(type_mapping)
df_valid = df_valid.astype(type_mapping)
df_test = df_test.astype(type_mapping)
```

```python
def get_hot_encoding(row, rel_cols):
    hot_encoding = [row[c] for c in rel_cols]
    return hot_encoding
```

```python
df_train["multi_labels"] = df_train.apply(lambda x: get_hot_encoding(x, list(type_mapping.keys())), axis=1)
df_valid["multi_labels"] = df_valid.apply(lambda x: get_hot_encoding(x, list(type_mapping.keys())), axis=1)
df_test["multi_labels"] = df_test.apply(lambda x: get_hot_encoding(x, list(type_mapping.keys())), axis=1)
```

```python
sentences_test = set(df_test["Sentence"].to_list())
sentences_train = set(df_train["Sentence"].to_list())
sentences_valid = set(df_valid["Sentence"].to_list())
sentences_valid_and_test = sentences_test.union(sentences_valid)

print(len(sentences_train & sentences_test))
print(len(sentences_train & sentences_valid))
print(len(sentences_test & sentences_valid))
```

```python
df_train_clean = df_train[~df_train["Sentence"].isin(sentences_valid_and_test)]
```
